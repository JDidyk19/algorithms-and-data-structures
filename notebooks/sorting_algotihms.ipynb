# Sorting Algorithm Comparison

```python
import matplotlib.pyplot as plt
from scipy.special import gamma
import numpy as np
```

### A graph of the time complexities of different algorithms

```python
n = np.linspace(1, 50)
plt.plot(n, n ** 2, label="bubble sort")
plt.plot(n, n + n, label="counting sort")
plt.plot(n, (n + 1) ** 2, label="insertion sort")
plt.plot(n, n * np.log(n), label="merge sort")
plt.plot(n, (n - 1) * np.log(n), label="quick sort")
plt.plot(n, (n - 1) ** 2, label="selection sort")
plt.legend(loc='upper left')
plt.show()
```

### A graph of the time complexity of Bogosort algorithm


```python
n = np.linspace(1, 10)
plt.plot(n, n * gamma(n), label="bogosort")
# plt.plot(n, n ** 5, label="bogosort")
plt.legend(loc='upper left')
plt.show()
```
